1. Escriba una funcion en Python que dada una tupla de elementos (numeros o letras) esta se
convierta a un string.

---Resultado de prueba <br>


```python
tupla = (1,2,3,4,5,6,7,8)
def  imprimetupla(valor):
    registro=''

    for  i in  valor:        
        registro=  registro + str(i)         
    return registro
        

print ("---------------")
print ("Tupla ingreada:" + str(tupla))
print ("resultado: " +str(imprimetupla(tupla)))
```
